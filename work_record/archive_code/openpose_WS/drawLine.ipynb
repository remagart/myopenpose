```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import numpy as np
import math

a = [[110,127],[160,182],[143,225],[121,173],[119,171],[115,167]]

temp1 = []
temp2 = []
for item in a:
    temp1.append(item[0])
    temp2.append(item[1])
    
print(temp1)
print(temp2)

plt.plot(temp1,temp2);
plt.scatter(temp1,temp2);
```

```python
b = [110,127]
temp1 = [b[0]+2,b[1]+2]
temp2 = [b[0]-2,b[1]+2]
temp3 = [b[0]-2,b[1]-2]
temp4 = [b[0]+2,b[1]-2]

tmp = []
bx = []
by = []

tmp.append(temp1)
tmp.append(temp2)
tmp.append(temp3)
tmp.append(temp4)

for item in tmp:
    bx.append(item[0])
    by.append(item[1])
bx.append(tmp[0][0])
by.append(tmp[0][1])

plt.plot(bx,by);
plt.scatter(bx,by);
```

```python
# a = [5,8]
# b = [1,11]
aa = {"x": 11,"y": 7}
bb = {"x":11, "y": 9}
direct = [bb["x"]-aa["x"],bb["y"]-aa["y"]]
print(direct)
value = math.sqrt(direct[0] * direct[0] + direct[1] * direct[1]) / 10
direct = [direct[0]/value,direct[1]/value]

print(direct)
dir1 = [direct[1]*-1,direct[0]*1]   # [-4,3]
dir2 = [direct[1]*1,direct[0]*-1]   # [4,-3]

print(dir1)
print(dir2)

tmp = [aa["x"],aa["x"]+dir1[0],bb["x"]+dir1[0],bb["x"],bb["x"]+dir2[0],aa["x"]+dir2[0],aa["x"]]
tmp2 = [aa["y"],aa["y"]+dir1[1],bb["y"]+dir1[1],bb["y"],bb["y"]+dir2[1],aa["y"]+dir2[1],aa["y"]]

# tmp = [5,1,9,12,8,4,1] #x
# tmp2 = [8,11,5,9,12,15,11] #y


plt.plot(tmp,tmp2);
plt.scatter(tmp,tmp2);
```
